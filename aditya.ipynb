1: Python program to enter a word amd check its palidrome

```python
# prompt: Python program to enter a word amd check its palidrome

word = input("Enter word: ")
reverse_word = word[::-1]
if word == reverse_word:
  print(f"The word {word} is a palindrome.")
else:
  print(f"The word {word} is not a palindrome.")
```

2: Python code for file handling

```python
# prompt: uploading the pdf tell amount of characters

!pip install PyPDF2
from google.colab import files
import io
import PyPDF2

uploaded = files.upload()

for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')

    # Read the PDF file from bytes
    pdf_reader = PyPDF2.PdfReader(io.BytesIO(uploaded[filename]))
    text = ""
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        text += page.extract_text()

    character_count = len(text)

    print(f"The total number of characters in the PDF is: {character_count}")
```

3:c.	Python code for image processing

```python
# prompt: only original, grayscale and rottated

!pip install opencv-python matplotlib
from google.colab import files
from PIL import Image
import io
import cv2
import matplotlib.pyplot as plt
import numpy as np

uploaded = files.upload()

for filename in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=filename, length=len(uploaded[filename])))
  img_pil = Image.open(io.BytesIO(uploaded[filename]))

  img_cv = cv2.cvtColor(np.array(img_pil), cv2.COLOR_RGB2BGR)

  gray_img = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)

  # Rotate the image
  rows, cols, _ = img_cv.shape
  M = cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0), 90, 1)
  rotated_img = cv2.warpAffine(img_cv, M, (cols,rows))


  # Create a figure with subplots for horizontal alignment
  fig, axes = plt.subplots(1, 3, figsize=(15, 5))

  axes[0].imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
  axes[0].set_title("Original Image")
  axes[0].axis('off')

  axes[1].imshow(gray_img, cmap='gray')
  axes[1].set_title("Grayscale Image")
  axes[1].axis('off')

  axes[2].imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
  axes[2].set_title("Rotated Image")
  axes[2].axis('off')


  plt.tight_layout()
  plt.show()
```

4:Python code for handling video files

```python

!pip install opencv-python
from google.colab import files
import cv2
import os

# Upload the video file
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')

    # Create a directory to save frames
    if not os.path.exists("frames"):
        os.makedirs("frames")

    # Read the video
    cap = cv2.VideoCapture(filename)
    frame_count = 0

    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret == False:
            break
        cv2.imwrite(f'frames/frame{frame_count}.jpg',frame)
        frame_count += 1

    cap.release()
    cv2.destroyAllWindows()
    print(f"Successfully extracted {frame_count} frames to the 'frames' directory.")
```
